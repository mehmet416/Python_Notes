## 1) Series

In Pandas, a Series is a one-dimensional labeled array that can hold data of various types. It's similar to a column in a spreadsheet or a one-dimensional array in NumPy, but with additional features that make data manipulation and analysis more convenient. Each element in a Series has an associated label called an index

**Creating a Series:**
You can create a Series by passing a list, array, or dictionary to the pd.Series() constructor.

```python
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series1 = pd.Series(data)
series2 = pd.Series(data, index=index)
print(series1)
print(series2)
# if you do not define any index, default values will replaced
```

```python
# To adjust the start value of series, use this method:

start_index_value = 1  # Set your desired start value
data = [10, 20, 30, 40, 50]

series = pd.Series(data, index=range(start_index_value, start_index_value + len(data)))
print(series)
```

**Accessing Elements:** You can access elements in a Series using index labels.

```python
print(series1[0])
print(series2['b'])  # Output: 20
```

**Basic Methods and Attributes:**

* values: Returns the values of the Series as a NumPy array.
* index: Returns the index labels of the Series.
* shape: Returns the dimensions of the Series.
* size: Returns the number of elements in the Series.

```python
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)
```

```python
series.index
```

```python
series.values
```

```python
series.size
```

**Filtering and Selection:**
You can use boolean indexing to filter and select elements in a Series.

```python
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)

filtered_series1 = series[series > 25]
filtered_series2 = series[(series > 25) & (series<40)]
print(filtered_series1)
print(filtered_series1)
```

**Common Methods:**

* head(n): Returns the first n elements of the Series.
* tail(n): Returns the last n elements of the Series.
* describe(): Generates summary statistics of the Series.
* mean(), .median(), .std(), .min(), .max(): Calculate various statistics of the Series.
* unique(), .nunique(): Get unique values and count of unique values.

```python
data = [10, 20, 30, 40, 50, 20, 30, 40]
index = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
series = pd.Series(data, index=index)
```

```python
# .head(n): Returns the first n elements of the Series
first_3_elements = series.head(3)
print("First 3 elements:")
print(first_3_elements)
```

```python
# .tail(n): Returns the last n elements of the Series
last_2_elements = series.tail(2)
print("\nLast 2 elements:")
print(last_2_elements)
```

```python
# .describe(): Generates summary statistics of the Series
summary_stats = series.describe()
print("\nSummary statistics:")
print(summary_stats)
```

```python
# .mean(), .median(), .std(), .min(), .max(): Calculate various statistics
mean_value = series.mean()
median_value = series.median()
std_deviation = series.std()
min_value = series.min()
max_value = series.max()
print("\nMean:", mean_value)
print("Median:", median_value)
print("Standard Deviation:", std_deviation)
print("Minimum Value:", min_value)
print("Maximum Value:", max_value)
```

```python
# .unique(), .nunique(): Get unique values and count of unique values
unique_values = series.unique()
num_unique_values = series.nunique()
print("\nUnique Values:", unique_values)
print("Number of Unique Values:", num_unique_values)
```

**Missing Data Handling:**
Pandas handles missing data effectively using NaN (Not a Number) values.

**Vectorized Operations:**
You can apply functions and operations element-wise to a Series without writing explicit loops.

```python
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series = pd.Series(data, index=index)

squared_series = series.apply(lambda x: x ** 2)
squared_series
```

* **parse_dates**: Columns to parse as datetime data.
* **na_values**: Values to treat as NaN (missing data).

**Creating a JSON File From a DataFrame**

The to_json() method in pandas is used to convert a DataFrame or Series to a JSON (JavaScript Object Notation) format. JSON is a popular data interchange format that is human-readable and can be easily processed by different programming languages. The to_json() method has several parameters, including orient and lines, that allow you to customize the structure of the JSON output. <br>
Here's what the orient and lines parameters do:


**1) orient:** <br>
The orient parameter specifies the format in which the JSON data is organized. It determines how the data is structured in the resulting JSON output. Some commonly used values for the orient parameter are:

* 'split': Dictionary with 'index', 'columns', and 'data' keys.
* 'records': List of dictionaries, each representing a row.
* 'index': Dictionary with keys as index values and values as rows.
* 'columns': Dictionary with keys as column names and values as columns.
* 'values': List of lists representing the values only.

**2) lines:** <br>
The lines parameter, if set to True, formats the JSON output with one record per line. This is useful for creating JSON files with a record on each line, which can be easier to read or process incrementally.

```python
import pandas as pd

# Sample DataFrame
data = {'Name': ['John', 'Anna', 'Peter'],
        'Age': [25, 30, 45],
        'City': ['New York', 'Paris', 'Berlin']}

df = pd.DataFrame(data)

# Using different orient values
json_split = df.to_json(orient='split')
json_records = df.to_json(orient='records')
json_index = df.to_json(orient='index')
json_columns = df.to_json(orient='columns')
json_values = df.to_json(orient='values')

print("JSON with 'split' orient:")
print(json_split)

print("\nJSON with 'records' orient:")
print(json_records)

print("\nJSON with 'index' orient:")
print(json_index)

print("\nJSON with 'columns' orient:")
print(json_columns)

print("\nJSON with 'values' orient:")
print(json_values)

# Using lines parameter
json_lines = df.to_json(orient='records', lines=True)
print("\nJSON with 'records' orient and lines=True:")
print(json_lines)
```

### Groupby Method
The groupby() method in pandas is used to group data from a DataFrame based on one or more columns and then apply aggregate functions to analyze or summarize the grouped data. This method is particularly useful for performing operations on specific groups of data within a larger datasetgroup.

**Here's how the groupby() method works:**

**Split:** The data is divided into groups based on the values in the specified column(s).

**Apply:** An aggregate function is applied to each group individually. This can be a function that calculates statistics like sum, mean, median, count, etc., or any custom function.

**Combine:** The results of the aggregation are combined into a new DataFrame or Series, where each row corresponds to a group.

You can use the groupby() method with multiple columns to group your data based on multiple criteria. When you pass a list of columns to the by parameter of the groupby() method, the data will be grouped by unique combinations of values in those columns.

```python
import pandas as pd

data = {
    'Product': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Region': ['North', 'North', 'South', 'North', 'North', 'South'],
    'Sales': [100, 200, 150, 250, 120, 180]
}

df = pd.DataFrame(data)

# Grouping by 'Product' and 'Region' columns, and calculating the sum of sales for each combination
grouped_product = df.groupby(['Product', 'Region'])
grouped_region = df.groupby(['Region', 'Product'])

total_sales1 = grouped_product['Sales'].sum()
total_sales2 = grouped_region['Sales'].sum()

print(total_sales1)
print()
print(total_sales2)
```

## 11) Concatenate, Join and Merge Functions 

The concat(), join(), and merge() functions in pandas are used to combine DataFrames or Series objects.

### 1) Concatenate Function
The concatenate method in pandas is used to combine two or more pandas objects (DataFrames or Series) along a particular axis. This function allows you to stack data vertically or horizontally, based on the specified axis. It's particularly useful when you have data split across multiple objects and want to consolidate them.

```python
import pandas as pd

# Create sample DataFrames
data1 = {
    'A' : ['A1', 'A2', 'A3', 'A4'],
    'B' : ['B1', 'B2', 'B3', 'B4'],
    'C' : ['C1', 'C2', 'C3', 'C4'],
}

data2 = {
    'A' : ['A5', 'A6', 'A7', 'A8'],
    'B' : ['B5', 'B6', 'B7', 'B8'],
    'C' : ['C5', 'C6', 'C7', 'C8'],
}

df1 = pd.DataFrame(data1, index=[1,2,3,4])
df2 = pd.DataFrame(data2, index=[5,6,7,8])
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])  # default value for  axis: 'Axis' = 0. if you do not define a certain axis, the function will defaultly add to in the direction of columns 
```

```python
pd.concat([df1,df2], axis=1)  # notice that since 1,2,3,4 indexes are nan values for df2,  also 5,6,7,8 indexes are  nan values for df1
```

```python
# if you want to concatenate df2 to df2 without nan values, change index numbers of either df1 or df2
df2.index = [1,2,3,4]
pd.concat([df1,df2], axis = 1)
```

### 2) Join Function
The join method in pandas is used to combine two or more DataFrame objects based on their indexes or on a specific column that is common between them. This function allows you to perform SQL-style joins, such as inner, outer, left, and right joins.

```python
data1 = {
    'A' : ['A1', 'A2', 'A3', 'A4'],
    'B' : ['B1', 'B2', 'B3', 'B4'],
    'C' : ['C1', 'C2', 'C3', 'C4'],
}

data2 = {
    'X' : ['X1', 'X2', 'X3'],
    'Y' : ['Y1', 'Y2', 'Y3'],
    
}

df1 = pd.DataFrame(data1, index=[1,2,3,4])
df2 = pd.DataFrame(data2, index=[1,2,3])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2) # df1 is the main set and if there any same index values of df2 with df1, it will be added to the rest of table as columns. 
# Notice that this method concatenate new columns to each other 
```

```python
df2.join(df1)
```

### 3) Merge Functon
The merge method in pandas is used to combine two or more DataFrame objects based on common columns or indexes. It provides more advanced and flexible merging capabilities compared to the join method. With merge, you can specify multiple columns to merge on, handle different types of joins, and customize how the merging operation is performed.

```python
data1 = {
    'A' : ['A1', 'A2', 'A3', 'A4'],
    'B' : ['B1', 'B2', 'B3', 'B4'],
    'Key': ['K1', 'K2', 'K3', 'K4']
}

data2 = {
    'X' : ['X1', 'X2', 'X3', 'X4'],
    'Y' : ['Y1', 'Y2', 'Y3', 'Y4'],
    'Key': ['K1', 'K3', 'K4', 'K3']

}

df1 = pd.DataFrame(data1, index=[1,2,3,4])
df2 = pd.DataFrame(data2, index=[1,2,3,4])
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on = 'Key')  # Merge function takes same values for both sets and then complete each of them with lost values. 
# Notice that order of indexes does not matter, any common values 'on stated' columns is enough.
# If there are more than one common columns, it is important to state the 'on' parameter which allows you to choose the searching column of common values.
```

```python
pd.merge(df2,df1, on = 'Key')
```
